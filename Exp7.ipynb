```python
# Reverse of a number

num = int(input("Enter a number:"))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //=10
print("Reversed number =", rev)
```

```python
# Check Palindrome number/string

num = int(input("Enter a number:"))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //=10
if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
st = input("Enter a string:")
rev =st[::-1]
if st==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
   print("Yes")
else:
    print("No")
```

```python
#Count digits in a number
num = int(input("Enter a number:"))
count = 0
while num > 0:
  count += 1
  num= num// 10

print("Number of digits =", count)
```

```python
nums = [10,20,30,40,50]
key = int(input("Enter element to search: "))
i = 0
while i < len(nums):
  if nums[i] == key:
    print("Element found at index", i)
    break

  i += 1

else:
  print("Element not found")
```
